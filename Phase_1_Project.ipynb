# Data Cleaning

This invloves dropping unnecessary data that doesnt support Kengeles Aviation decision to purchase aircrafts.


```python
#dropping the unncessary columns in whole as this research mainly focuses on finding out what are best aircraft with the least likelihood of causing an accident

relevant_colums = [
    'Event.Id',
    'Investigation.Type',
    'Injury.Severity',
    'Aircraft.damage',
    'Aircraft.Category',
    'Make',
    'Model',
    'Purpose.of.flight',
    'Air.carrier',
    'Total.Fatal.Injuries',
    'Total.Serious.Injuries',
    'Total.Minor.Injuries',
    'Total.Uninjured'
]
```

```python
df = df [relevant_colums]
```

```python
df.shape
```

```python
df.head()
```

```python
#mainly focusing on the accident type of investigation as incidents are assumed to be caused by human error
df =df[df["Investigation.Type"] == "Accident"]
df
```

```python
df.shape
```

```python
#droping all null values to reduce the data set
df.dropna (inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
# Filter for aircraft category 'Airplane'

df = df[df['Aircraft.Category'] == 'Airplane']
df.shape
```

```python
df['Purpose.of.flight'].unique()
```

```python
#We will only use planes who's purpose of flight include Business, and Executive as they align with the business needs.
df = df[df['Purpose.of.flight'].isin(['Business', 'Executive/corporate'])]
df.shape
```

#Key Business

This is where the data has been cleaned to mainly contain what is relevant for the business


```python
# 5 aircrafts that have the highest fatal injuries

top_fatal_aircraft = df.groupby(['Make', 'Model'])['Total.Fatal.Injuries'].sum().nlargest(5)
print(top_fatal_aircraft)
```

```python
# A graph containing the 5 most dangerous planes makes and model

top_fatal_aircraft.plot(kind='bar', figsize=(15, 10))
plt.title('Top 5 Aircraft with Highest Fatal Injuries')
plt.xlabel('Aircraft Make and Model')
plt.ylabel('Total Fatal Injuries')
plt.xticks(rotation=45)
plt.show()
```

```python
 #The 10 aircrafts make and model that have the least fatal injuries

 least_fatal_aircraft = df.groupby(['Make','Model'])['Total.Fatal.Injuries'].sum().nsmallest(10)
 print(least_fatal_aircraft)
```

```python
# TOP 10 MAKE AND MODEL WITH HIGHEST TOTAL UNINJURED

top_uninjured_aircraft = df.groupby(['Make', 'Model'])['Total.Uninjured'].sum().nlargest(10)
print(top_uninjured_aircraft)


```

```python
# 5 MAKES WITH MOST AIRCRAFT DAMAGE

top_damage_makes = df.groupby('Make')['Aircraft.damage'].count().nlargest(5)
print(top_damage_makes)
```

```python
#5 MAKES WITH LEAST AIRCRAFT DAMAGE

least_damage_makes = df.groupby('Make')['Aircraft.damage'].count().nsmallest(10)
print(least_damage_makes)
```

```python
#GRAPH WITH LEAST AIRCRAFT DAMAGE

least_damage_makes.plot(kind='bar', figsize=(15, 10))
plt.title('10 Aircraft Makes with Least Aircraft Damage')
plt.xlabel('Aircraft Make')
plt.ylabel('Number of Aircraft Damage')
plt.show()
```

```python

# Filter for AERO COMMANDER aircraft
aero_commander_df = df[df['Make'] == 'AERO COMMANDER']

# Get the top 5 models with the highest count
top_aero_commander_models = aero_commander_df.groupby('Model')['Model'].count().nlargest(5)
print(top_aero_commander_models)
```
